```python
import polars as pl
import plotly.express as px

# Магия путей: '../' означает "выйти из папки notebooks на один уровень выше"
file_path = "../data/raw/sensor_geometry.csv"

# Загружаем данные
df = pl.read_csv(file_path)

# Строим 3D график
fig = px.scatter_3d(
    df.to_pandas(), 
    x='x', y='y', z='z',
    opacity=0.8,
    title='IceCube 3D Geometry',
    template='plotly_dark'
)
fig.update_traces(marker=dict(size=2))
fig.show()
```
